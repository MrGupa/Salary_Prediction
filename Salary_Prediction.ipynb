```python
#Filtering the data for exp

def filter_exp(x):
  if x ==  'More than 50 years':
    return 50
  if x == 'Less than 1 year':
    return 0.5
  return float(x)
```

```python
filter_exp('More than 50 years')
```

```python
df['YearsCodePro'] = df['YearsCodePro'].apply(filter_exp)
```

```python
df['YearsCodePro'].unique()
```

```python
df.sample(5)
```

```python
df['EdLevel'].unique()
```

```python
def filter_education(x):
  if 'Bachelor’s degree' in x:
    return 'Bachelor’s degree'
  if 'Master’s degree' in x:
    return 'Master’s degree'
  if 'Professional degree' in x or 'Other doctoral' in x:
    return 'Post grad'
  return 'Less than a Bachelors'
```

```python
filter_education('I never completed any formal education')
```

```python
df['EdLevel'] = df['EdLevel'].apply(filter_education)
```

```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np



# def load_model():
#   with open('deployment.pkl','rb') as file:
#     data = pickle.load(file)
#   return data

pickle_in = open('deployment.pkl','rb')
data = pickle.load(pickle_in)
# data = load_model()

model = data['model']  #Algo
le_country = data['le_country']   #Label Encoder
le_education = data['le_education']


st.title("Salary Prediction")



countries = (
    'United States',
    'India',
    'United Kingdom',
    'Germany',
    'Canada',
    'Brazil',
    'France',
    'Spain',
    'Netherlands',
)

education = (
    'Less than a Bachelors',
    'Bachelor’s degree',
    'Master’s degree',
    'Post grad',
)

country = st.selectbox("Country",countries)
education = st.selectbox("Education Level",education)
experience = st.slider("Years of Experience",0,50)

ok = st.button("Calculate Salary")
if ok:
  X = np.array([[country,education,experience]])
  X[:,0] = le_country.transform(X[:,0])
  X[:,1] = le_education.transform(X[:,1])
  X = X.astype(float)
  salary = model.predict(X)
  st.subheader(f"The estimated salary is ${salary[0]:.2f}")
```
